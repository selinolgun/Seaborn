## Histogram ve Yoğunluk Grafikleri

```python
iris = pd.read_csv("iris.data")
iris.head()
```

### Veri setimizi okuttuk.
### Veri setimizin içeriği hakkında bilgi sahibi olmak için ilk 5 satırını `head()` metoduyla yazdırdık.

```python
setosa = iris.loc[iris.Turu == "Iris-setosa"]
virginica = iris.loc[iris.Turu == "Iris-virginica"]
```

### Tür için değişkenlerimizi `setosa` ve `virginica` olarak oluşturduk.

```python
setosa.Canak_yaprak_boyu.plot.hist()
```

### `Setosa` türünün `çanak yaprak boyu`nu `histogram grafiği` haline getirmiş olduk.

```python
sns.kdeplot(setosa.Canak_yaprak_boyu, shade= True, color= "b")
```

### `kdeplot()` metoduyla Setosa türünün çanak yaprak boyunun yoğunluk grafiğini oluşturduk.

```python
sns.distplot(iris.Canak_yaprak_boyu)
```

### Çoğunlukla histogram ve yoğunluk grafiği aynı anda istenir.
### Bu yüzden `distplot()` metoduyla ikisini aynı grafik içinde görebiliyoruz.

```python
sns.distplot(iris.Canak_yaprak_boyu)
sns.distplot(iris.Canak_yaprak_eni)
```

### Art arda iki kod yazarsak aynı grafik içinde iki grafik birden görülebilir.

## Facet Histogramlar

```python
bahsis = sns.load_dataset("tips")
bahsis.head()
```

### Veri setimizi okuttuk.

```python
bahsis["bahsis_yuzde"] = bahsis["tip"] * 100 / bahsis["total_bill"]
bahsis.head()
```

### Bahşiş yüzdesini hesaplayıp onunla yeni bir sütun oluşturduk.

```python
grid = sns.FacetGrid(bahsis, col= "time", row= "smoker", margin_titles= True)
grid.map(plt.hist, "bahsis_yuzde", bins= np.linspace(0,40, 15))
```

### `FacedGrid` özelliği bize grafikleri ayrı ayrı incelememize olanak tanır.
